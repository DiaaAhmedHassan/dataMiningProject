## Data Cleaning

*as we see here, there is no null values in the data set*

```python
to_drop = ['id', 'waterfront', 'zipcode', 'date', 'yr_renovated']
edited_feature_houses = original_houses_data.drop(to_drop, axis=1)
original_houses_data.head(11)
```

```python
# remove duplicates
houses_no_duplicates = edited_feature_houses.drop_duplicates()
houses_no_duplicates.shape
```

```python
# fix some columns
columns_list = ['bedrooms', 'bathrooms', 'floors']
unique_values = houses_no_duplicates[columns_list].apply(pd.Series.unique)
unique_values
```

*As we see here, there are some logical mistakes regarding <br>
the number of floors and bathrooms, we need to fix that*

```python
columns_list = ['bathrooms', 'floors']
houses = houses_no_duplicates.copy()
houses[columns_list] = houses_no_duplicates[columns_list].astype(int)
houses.head()
```

## Data exploration <br>
start with plotting the data of the houses according to longitude and latitude to see if there is any visual patterns 

```python
import matplotlib.pyplot as plt

houses.plot(kind='scatter', x = "long", y='lat', s=(houses['sqft_living'])/100, label='House price', c='price', cmap=plt.get_cmap("jet"))
```

# Correlation matrix

now we try to find the correlations between deferent features

```python
correlation_matrix = houses.corr()
correlation_matrix.style.background_gradient(cmap='jet')
```

## plot the strongly correlated data

```python
plt.subplot(3,2,1)
plt.scatter(x=houses['price'], y=houses['sqft_living'])
plt.xlabel('price')
plt.ylabel('sqft_living')

plt.subplot(3,2,2)
plt.scatter(x=houses['sqft_living'], y=houses['sqft_living15'])
plt.xlabel('sqft_living')
plt.ylabel('sqft_living15')

plt.subplot(3,2,3)
plt.scatter(x=houses['sqft_living'], y=houses['sqft_above'])
plt.xlabel('sqft_living')
plt.ylabel('sqft_above')

plt.subplot(3,2,4)
plt.scatter(x=houses['sqft_lot'], y=houses['sqft_lot15'])
plt.xlabel('sqft_lot')
plt.ylabel('sqft_lot15')

plt.subplot(3,2,5)
plt.scatter(x=houses['grade'], y=houses['sqft_above'])
plt.xlabel('grade')
plt.ylabel('sqft_above')

plt.tight_layout()
```
